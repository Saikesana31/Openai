```python
# Here we define the third part data through api call(weather data)

import requests
import os
from dotenv import load_dotenv
from datetime import datetime

def current_weather(location):
    url = "https://ai-weather-by-meteosource.p.rapidapi.com/time_machine"

    # Extract city name from location string (e.g., "Delhi, India" -> "delhi")
    # Remove country/state info and convert to lowercase
    place_id = location.split(',')[0].strip().lower()
    today = datetime.now().strftime("%Y-%m-%d")

    querystring = {"place_id": place_id, "date": today, "units": "auto"}

    headers = {
	    "x-rapidapi-key": "7d146ca364mshbfc4ae9d1918d24p13f8fejsn45e1c216da91",
	    "x-rapidapi-host": "ai-weather-by-meteosource.p.rapidapi.com"
    }

    try:
        response = requests.get(url, headers=headers, params=querystring)
        response.raise_for_status()  # Raises an HTTPError for bad responses
        
        data = response.json()
        print(data)
        return data
    except requests.exceptions.HTTPError as e:
        # Handle HTTP errors (4xx, 5xx)
        error_data = response.json() if response.text else {"error": str(e)}
        print(f"HTTP Error {response.status_code}: {error_data}")
        return error_data
    except requests.exceptions.RequestException as e:
        # Handle network errors, timeouts, etc.
        print(f"Request error: {e}")
        return {"error": str(e)}
    except ValueError as e:
        # JSON decode error
        print(f"Error parsing response: {e}")
        return {"error": f"Invalid response: {response.text[:200]}"}
```

```python
# define function
import openai
import json

functions = [
    {
        "name": "current_weather",
        "description": "Get the current weather in a given location",
        "parameters": {
            "type": "object",
            "properties": {
                "location": {
                    "type": "string",
                    "description": "The city and state, e.g. San Francisco, CA or Delhi, India"
                }
            },
            "required": ["location"]
        }
    }
]
```

```python
# Make a chat completion with function calling
user ="hi there"
messages =[]
messages.append({"role": "user", "content": user})

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=messages,
    functions=functions,
    function_call="auto"  # "auto" means the model can choose to call a function
)

print(response)
```

```python
k = response.choices[0].message["function_call"]["arguments"]
print(k)
print(type(k))
```

```python
# Here we connect the function with the model

message = response.choices[0].message

if message.get("function_call"):
    # The model wants to call a function
    function_name = message["function_call"]["name"]
    function_args = json.loads(message["function_call"]["arguments"])

    print(f"Function to call: {function_name}")
    print(f"Arguments: {function_args}")

     # Call the actual function
    if function_name == "current_weather":
        function_response = current_weather(function_args.get("location")) #weather data (likely a dict)

        # Add the function response to messages
        messages.append(message)  # Add assistant's message with function call
        messages.append({
            "role": "function",
            "name": function_name,
            "content": json.dumps(function_response)
        })

        # Get the final response from the model
        second_response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            functions=functions
        )
        
        print("\nFinal response:")
        print(second_response.choices[0].message.content)
else:
    # No function call needed
    print(message.content)
```
